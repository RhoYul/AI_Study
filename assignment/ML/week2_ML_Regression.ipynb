```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('C:/Users/USER/Downloads/dataset.csv', index_col=0)
df
```

```python
X = df.drop('trestbps',axis=1)
y = df['trestbps']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#데이터 전처리
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
svr_model = SVR()
svr_model.fit(X_train_scaled, y_train)
pred_svr = svr_model.predict(X_test_scaled)
print("--------------------Predict-----------------\n", pred_svr)
print(svr_model.score(X_train_scaled, y_train))
mse_svr = np.sqrt(mean_squared_error(pred_svr, y_test))
print('평균제곱근오차', mse_svr)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train)
pred_lr = lr_model.predict(X_test_scaled)
print("--------------------Predict-----------------\n", pred_lr)
print(lr_model.score(X_train_scaled, y_train))
mse_lr = np.sqrt(mean_squared_error(pred_lr, y_test))
print('평균제곱근오차', mse_lr)
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
pred_dt = dt_model.predict(X_test)
print("--------------------Predict-----------------\n", pred_dt)
print(dt_model.score(X_train, y_train))
mse_dt = np.sqrt(mean_squared_error(pred_dt, y_test))
print('평균제곱근오차', mse_dt)
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
pred_rf = rf_model.predict(X_test)
print("--------------------Predict-----------------\n", pred_rf)
print(rf_model.score(X_train, y_train))
mse_rf = np.sqrt(mean_squared_error(pred_rf, y_test))
print('평균제곱근오차', mse_rf)
```

```python
knr_model = KNeighborsRegressor()
knr_model.fit(X_train, y_train)
pred_knr = knr_model.predict(X_test)
print("--------------------Predict-----------------\n", pred_knr)
print(knr_model.score(X_train, y_train))
mse_knr = np.sqrt(mean_squared_error(y_test, pred_knr))
print('평균제곱근오차', mse_knr)
```
