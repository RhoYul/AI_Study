```python
df = df.dropna()
df = df.drop(['Name', 'Ticket'], axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
label = LabelEncoder()
df['Embarked'] = label.fit_transform(df['Embarked'])
df['Sex'] = label.fit_transform(df['Sex'])
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#상관 관계 행렬
data_corr=df.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(data_corr, annot=True, cbar=False)
plt.show()
```

```python
X = df.drop(['Survived'], axis=1)
y = df['Survived']
X
```

```python
# train/test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#데이터 전처리
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
rs.fit(X_train)
X_train_scaled = rs.transform(X_train)
X_test_scaled = rs.transform(X_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train_scaled)
X_poly_test = poly.fit_transform(X_test_scaled)
```

```python
# SVM

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(X_train_scaled, y_train)

pred_svm = clf_svm.predict(X_test_scaled)

print("\n--- SVM Classifier ---")
print(clf_svm.score(X_train_scaled, y_train))
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
# LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(C=100, max_iter=5000)
clf_lr.fit(X_poly_train, y_train)

pred_lr = clf_lr.predict(X_poly_test)

print ("\n--- Logistic Regression Classifier ---")
print(clf_lr.score(X_poly_train, y_train))
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
params = {'max_depth': randint(20,60),
          'min_impurity_decrease': uniform(0.0001, 0.001),
          'min_samples_split': randint(2, 25),
          'min_samples_leaf': randint(1, 25)}
gs = RandomizedSearchCV(DecisionTreeClassifier(random_state=0), params, n_iter=100, n_jobs=-1,random_state=42)
gs.fit(X_poly_train, y_train)
print(gs.best_params_)
print(np.max(gs.cv_results_['mean_test_score']))
```

```python
# DT

from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(max_depth=58, min_impurity_decrease=0.0007510770255019445, 
                                min_samples_leaf=7, min_samples_split=10)
clf_dt.fit(X_poly_train, y_train)

pred_dt = clf_dt.predict(X_poly_test)

print ("\n--- Decision Tree Classifier ---")
print (clf_dt.score(X_poly_train, y_train))
print (clf_dt.score(X_poly_test, y_test))
print (confusion_matrix(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
params = {'n_estimators': randint(80, 120),
          'min_impurity_decrease': uniform(0.0001, 0.001),
          'min_samples_split': randint(2, 25),
          'min_samples_leaf': randint(1, 25)}
gs = RandomizedSearchCV(RandomForestClassifier(random_state=0), params, n_iter=100, n_jobs=-1,random_state=42)
gs.fit(X_poly_train, y_train)
print(gs.best_params_)
print(np.max(gs.cv_results_['mean_test_score']))
```

```python
# RT

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(min_impurity_decrease=0.00028223608778806236, min_samples_leaf=4,
                                min_samples_split=3, n_estimators=85)
rf_clf.fit(X_poly_train, y_train)
pred = rf_clf.predict(X_poly_test)
print (rf_clf.score(X_poly_train, y_train))
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test, pred))
```
